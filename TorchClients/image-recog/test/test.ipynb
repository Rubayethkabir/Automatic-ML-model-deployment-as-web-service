```python
from IPython.display import Image
from comm import Client
```

```python
### Service should be started on the server as:
### $ nvidia-docker run -it --rm -p 10010:10000 imagerecog_service

#Initialize the Client
cl = Client()
```

```python
# Initialize the service
cl.connectService('dl-wkr6.u-aizu.ac.jp', 10010)
cl.initService()
```

```python
# Get Confirmation
reply = cl.receiveData()
for command, data_type, data in reply:
    print(command)
```

```python
# Read initial image 
#orig_img = open('elephant.jpg', 'rb').read()
orig_img = open('car.jpg', 'rb').read()

# Send request
cl.connectService('dl-wkr6.u-aizu.ac.jp', 10010)
cl.sendImageRequest(orig_img)

# Show request
Image(orig_img)
```

```python
# Display result
reply = cl.receiveData()
for command, data_type, data in reply:
    print(data)
```

```python
# Stop Service
cl.connectService('dl-wkr6.u-aizu.ac.jp', 10010)
cl.stopService()
cl.disconnectService()
```

```python
from IPython.display import Image, display
from comm import Client
```

```python
### Service should be started on the server as:
### $ nvidia-docker run -it --rm -p 10010:10000 styletransf_service

# Server the service is running
server = 'dl-wkr6.u-aizu.ac.jp'
#Initialize the Client
cl = Client()
```

```python
# Initialize the service
# Load style image
style_img = open('../styles/wave.jpg', 'rb').read()
cl.connectService(server, 10010)
cl.initService(style_img, 'image')

# Show style
Image(style_img)
```

```python
# Get Confirmation
reply = cl.receiveData()
for command, data_type, data in reply:
    print(command)
```

```python
# Read initial image 
orig_img = open('chicago.jpg', 'rb').read()

# Send request
cl.connectService(server, 10010)
cl.sendImageRequest(orig_img)

# Show request
Image(orig_img)
```

```python
# Display result
reply = cl.receiveData()
for command, data_type, data in reply:
    display(Image(data))
```

```python
# Stop Service
cl.connectService(server, 10010)
cl.stopService()
cl.disconnectService()
```
